```python
# carga y umbralizacion de flechas que no sirven
picList = os.listdir('img/nossirve')
i = 0
imgListNossirve = []
for pic in picList:
    pic = "img/nossirve/" + pic
    img = cv2.imread(pic)
    img = cv2.resize(img, (1200, 900), interpolation= cv2.INTER_LINEAR)
    width = img.shape[1]
    height = img.shape[0]
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    threshType, threshImg = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY)
    filteredImg = cv2.bilateralFilter(threshImg, 5, 80, 80)
    imgListNossirve.append(threshImg)
    x = 0
    y = 0
    yfin = 0
    colSuma = np.sum(imgListNossirve[i], 0)
    for j in range(len(colSuma)) :
        if colSuma[j] < 255*(height-10) :
            x = j
            break
    filSuma = np.sum(imgListNossirve[i], 1)
    for k in range(len(filSuma)) :
        if filSuma[k] < 255*(height-1) :
            y = k
            break
    for l in reversed(range(len(filSuma))) :
        if filSuma[l] < 255*(height-1) :
            yfin = l
            break
    roi = [x, y-75, 500, 200]
    imgListNossirve[i] = imgListNossirve[i][int(roi[1]):int(roi[1]+roi[3]), int(roi[0]):int(roi[0]+roi[2])]
    plt.imshow(imgListNossirve[i], cmap = plt.cm.gray)
    plt.show()
    i = i+1
```

```python
# carga y umbralizacion de flechas que sirven
picList = os.listdir('img/sirve')
i = 0
imgListSirve = []
for pic in picList:
    pic = "img/sirve/" + pic
    img = cv2.imread(pic)
    img = cv2.resize(img, (1200, 900), interpolation= cv2.INTER_LINEAR)
    width = img.shape[1]
    height = img.shape[0]
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    threshType, threshImg = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    filteredImg = cv2.bilateralFilter(threshImg, 5, 80, 80)
    imgListSirve.append(threshImg)
    x = 0
    y = 0
    colSuma = np.sum(imgListSirve[i], 0)
    for j in range(len(colSuma)) :
        if colSuma[j] < 255*(height-10) :
            x = j
            break
    filSuma = np.sum(imgListSirve[i], 1)
    filSumaInv = np.sum(cv2.bitwise_not(imgListSirve[i]), 1)
    for k in range(len(filSuma)) :
        if filSuma[k] < 255*(height-1) :
            y = k
            break
    for l in reversed(range(len(filSuma))) :
        if filSuma[l] < 255*(height-1) :
            yfin = l
            break
    roi = [x, y-80, 500, 200]
    imgListSirve[i] = imgListSirve[i][int(roi[1]):int(roi[1]+roi[3]), int(roi[0]):int(roi[0]+roi[2])]
    plt.imshow(imgListSirve[i], cmap = plt.cm.gray)
    plt.show()
    i = i+1
```

```python
# deteccion de bordes flechas que no sirven
i = 0
edgeListNossirve = []
hNo = []
for image in imgListNossirve:
    edge = cv2.Canny(image, 100,200)
    kernel = np.ones((5,5),np.uint8)
    edge = cv2.dilate(edge, kernel, iterations = 1)
    edge = cv2.medianBlur(edge, 5)
    edge = cv2.morphologyEx(edge, cv2.MORPH_OPEN, kernel)
    edgeListNossirve.append(cv2.bitwise_not(edge))
    width = edgeListNossirve[i].shape[1]
    height = edgeListNossirve[i].shape[0]
    x = 0
    y = 0
    colSuma = np.sum(edgeListNossirve[i], 0)
    for j in range(len(colSuma)) :
        if colSuma[j] < 255*(height-10) :
            x = j
            break
    filSuma = np.sum(edgeListNossirve[i], 1)
    for k in range(len(filSuma)) :
        for kk in range(width):
            if edgeListNossirve[i][k][kk] == 0 :
                y = k
                break
    for l in reversed(range(len(filSuma))) :
        for ll in range(width):
            if edgeListNossirve[i][l][ll] == 0 :
                yfin = l
                break
    roi = [x, yfin, 400, y-yfin]
    hNo.append(y-yfin)
    edgeListNossirve[i] = edgeListNossirve[i][int(roi[1]):int(roi[1]+roi[3]), int(roi[0]):int(roi[0]+roi[2])]
    plt.imshow(edgeListNossirve[i], cmap = plt.cm.gray)
    plt.show()
    i = i+1
```

```python
# deteccion de bordes flechas que sirven
i = 0
edgeListSirve = []
hSi = []
for image in imgListSirve:
    edge = cv2.Canny(image, 100,200)
    kernel = np.ones((5,5),np.uint8)
    edge = cv2.dilate(edge, kernel, iterations = 1)
    edge = cv2.medianBlur(edge, 5)
    edge = cv2.morphologyEx(edge, cv2.MORPH_OPEN, kernel)
    edgeListSirve.append(cv2.bitwise_not(edge))
    width = edgeListSirve[i].shape[1]
    height = edgeListSirve[i].shape[0]
    x = 0
    y = 0
    colSuma = np.sum(edgeListSirve[i], 0)
    for j in range(len(colSuma)) :
        if colSuma[j] < 255*(height-10) :
            x = j
            break
    filSuma = np.sum(edgeListSirve[i], 1)
    for k in range(len(filSuma)) :
        for kk in range(width):
            if edgeListSirve[i][k][kk] == 0 :
                y = k
                break
    for l in reversed(range(len(filSuma))) :
        for ll in range(width):
            if edgeListSirve[i][l][ll] == 0 :
                yfin = l
                break
    roi = [x, yfin, 400, y-yfin]
    hSi.append(y-yfin)
    edgeListSirve[i] = edgeListSirve[i][int(roi[1]):int(roi[1]+roi[3]), int(roi[0]):int(roi[0]+roi[2])]
    plt.imshow(edgeListSirve[i], cmap = plt.cm.gray)
    plt.show()
    i = i+1
```

```python
# flood fill de las que no sirven
i = 0
floodListNossirve = []
sumNo = []
for image in edgeListNossirve:
    im_floodfill = image.copy()
    h, w = image.shape[:2]
    mask = np.zeros((h+2, w+2), np.uint8)
    #print(h, w)
    seedH = int((h+1)/2)
    cv2.floodFill(im_floodfill, mask, (200, seedH), 0)
    cv2.floodFill(im_floodfill, mask, (200, seedH+5), 0)
    cv2.floodFill(im_floodfill, mask, (200, seedH-5), 0)
    cv2.floodFill(im_floodfill, mask, (260, seedH), 0)
    cv2.floodFill(im_floodfill, mask, (160, seedH), 0)
    floodListNossirve.append(im_floodfill)
    plt.imshow(im_floodfill, cmap = plt.cm.gray)
    plt.show()
        
    sumNo.append(0)
    for j in range(0,h):
        for jj in range(0,w):
            if floodListNossirve[i][j][jj] == 0:
                sumNo[i] = sumNo[i] + 1
    i = i+1
```

```python
# flood fill de las que si sirven
i = 0
floodListSirve = []
sumSi = []
for image in edgeListSirve:
    if type(image) == tuple:
        im_floodfill = image[1].copy()
        h, w = image[1].shape[:2]
    else:
        im_floodfill = image.copy()
        h, w = image.shape[:2]
    mask = np.zeros((h+2, w+2), np.uint8)
    #print(h, w)
    seedH = int((h+1)/2)
    cv2.floodFill(im_floodfill, mask, (200, seedH), 0)
    cv2.floodFill(im_floodfill, mask, (200, seedH+5), 0)
    cv2.floodFill(im_floodfill, mask, (200, seedH-5), 0)
    cv2.floodFill(im_floodfill, mask, (260, seedH), 0)
    cv2.floodFill(im_floodfill, mask, (160, seedH), 0)
    cv2.floodFill(im_floodfill, mask, (30, seedH), 0)
    floodListSirve.append(im_floodfill)
    plt.imshow(im_floodfill, cmap = plt.cm.gray)
    plt.show()
    
    sumSi.append(0)
    for j in range(0,h):
        for jj in range(0,w):
            if floodListSirve[i][j][jj] == 0:
                sumSi[i] = sumSi[i] + 1
    i = i+1
```

```python
##### PERCEPTRON
vectores = []
y = []
features = 2
#x, y = make_blobs(n_samples=N, n_features=features, centers=2, cluster_std=2, random_state=2)
for i in range(len(hNo)):
    vectores.append([1, hNo[i], sumNo[i]])
    y.append(0)
for i in range(len(hSi)):
    vectores.append([1, hSi[i], sumSi[i]])
    y.append(1)
N = len(vectores)
print(vectores, y)
xext = vectores
k = 1000
w = np.random.rand(features+1)
print("Pesos iniciales", w)
message = "Mensaje"
yResult = []
lrate = 0.2
for iteration in range(k):
    errorCheck = False
    yTemp = []
    for i in range(N):
        net = np.dot(xext[i], w)
        if (net >= 0):
            yTemp.append(1)
        else:
            yTemp.append(0)
        if (iteration == k-1):
            print(y[i], yTemp[i])
        error = y[i] - yTemp[i]
        if(error != 0):
            errorCheck = True
            w[0] = w[0] + error * xext[i][0] * lrate
            w[1] = w[1] + error * xext[i][1] * lrate
            w[2] = w[2] + error * xext[i][2] * lrate
        yResult = np.asarray(yTemp)
    if (not errorCheck):
        print("Todos los errores son cero")
        print("Última iteración: ", iteration)
        break
    if (iteration == k-1):
        print("Límite de iteraciones alcanzado")
        break

print("Pesos resultantes: ", w)
```
